```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Hotel Bookings.csv")
```

```python
df.fillna(np.nan,inplace=True)
```

```python
df.head()
```

```python
count=df['hotel'].value_counts()
ax=sns.barplot(x=count.index,y=count.values)
plt.title("Number of times the hotels booked")
plt.xlabel("Count",size=12)
plt.ylabel("Hotels",size=12)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.histplot(df['lead_time'],bins=20)
plt.title("Lead time distribution")
plt.xlabel("Lead",size=12)
plt.ylabel("Frequency",size=12)
plt.show()
```

```python
df['customer_type'].unique()
```

```python
count=df['customer_type'].value_counts()
ax=sns.barplot(x=count.index,y=count.values)
plt.title("Types of customers booking the hotel")
plt.xlabel("Customer Types",size=12)
plt.ylabel("Count",size=12)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
grouped_data = df.groupby('hotel')['stays_in_week_nights'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(grouped_data['hotel'], grouped_data['stays_in_week_nights'], color=['skyblue', 'orange'])
plt.xlabel('Hotel')
plt.ylabel('Average Stays in Week Nights')
plt.title('Comparison of Stays in Week Nights by Hotel Type')
plt.show()
```

```python
grouped_data = df.groupby('hotel')['stays_in_weekend_nights'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(grouped_data['hotel'], grouped_data['stays_in_weekend_nights'], color=['skyblue', 'orange'])
plt.xlabel('Hotel')
plt.ylabel('Average Stays in Weekend Nights')
plt.title('Comparison of Stays in Weekend Nights by Hotel Type')
plt.show()
```

```python
grouped_data = df.groupby('hotel')['previous_cancellations'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(grouped_data['hotel'], grouped_data['previous_cancellations'], color=['skyblue', 'orange'])
plt.xlabel('Hotel')
plt.ylabel('previous_cancellations')
plt.title('Comparison of Stays in previous_cancellations by Hotel Type')
plt.show()
```

```python
grouped_data = df.groupby('hotel')['arrival_date_week_number'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(grouped_data['hotel'], grouped_data['arrival_date_week_number'], color=['skyblue', 'orange'])
plt.xlabel('Hotel')
plt.ylabel('arrival_date_week_number')
plt.title('Comparison of Stays in arrival_date_week_number by Hotel Type')
plt.show()
```

```python
group=df['arrival_date_month'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.plot(group.index,group.values,"o-")
plt.xticks(rotation=30)
plt.xlabel('Months')
plt.ylabel('Count')
plt.title('Visualizing Month that has highest bookings')
plt.grid()
plt.show()
```

```python
mean_lead_time = df.groupby('market_segment')['lead_time'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
mean_lead_time.plot(kind='bar', color='skyblue')
plt.xlabel('Market Segment')
plt.ylabel('Mean Lead Time')
plt.title('Mean Lead Time Comparison by Market Segment')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='hotel', y='lead_time', data=df)
plt.xlabel('Hotel')
plt.ylabel('Lead Time')
plt.title('Lead Time Comparison between Hotels')
plt.show()
```

```python
numerical_columns = df[['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'adr']]
sns.pairplot(numerical_columns)
plt.title('Pairplot of Numerical Columns')
plt.show()
```

```python
numerical_columns = df[['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'adr']]

# Calculating correlation matrix
correlation_matrix = numerical_columns.corr()

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Numerical Columns')
plt.show()
```
